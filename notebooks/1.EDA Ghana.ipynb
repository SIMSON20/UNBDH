## Optimization

For every market, let be:
+ Indices
    
    - i: commodity per category, $\in \{1, ..., 10\}$
    - j: category number, $\in \{1, ..., 4\}$

+ Variables

    - $A_{ij}$: Quantity of commodity i from category j. 
    

+ Parameters
    - $P_{ij}$: Price of commodity i from category j.
    - $M_{j}$: Maximum quantity of commodities from category j to comsume per month
    - $B$: Mensual budget of the consumer

+ Objective

    Min $\sum_{i=1}^{10}\sum_{j=1}^{4}P_{ij}\times A_{ij}$

+ Constraints

    $\sum_{i=1}^{10}A_{ij} \geq M_{j}$, $\forall j$ 
    
    $\sum_{i=1}^{10}\sum_{j=1}^{4}P_{ij}\times A_{ij} \leq B$

```python
consumption = gp.Model("Month Consumption Planning")

aij = consumption.addVars(commodities, categories, name = "aij") # quantity consummed
```

```python
# categories minimum levels
mlbycats = consumption.addConstrs((gp.quicksum(aij[commodity, category] 
                 for commodity in commodities) >= mjs[category]
                             for category in categories
#                                     if commodity in accra.loc[accra.category==category].commodity.unique().tolist()
                             ), name="mlbycats")

# zeroes
zeroes = consumption.addConstrs(
(aij[com, cat]==0 
 for com in commodities for cat in categories 
 if com not in accra.loc[accra.category==cat].commodity.unique().tolist()), name="zeroes")

# nozeroes
# nozeroes = consumption.addConstrs(
# (aij[com, cat]>=0 
#  for com in commodities for cat in categories 
#  if com in accra.loc[accra.category==cat].commodity.unique().tolist()), name="nozeroes")

# budget capacity
budget = consumption.addConstr(gp.quicksum(aij[commodity, category]*prices[commodity, category]
                                       for commodity in commodities for category in categories
                                          if  commodity in accra.loc[accra.category==category].commodity.unique().tolist())  <= budg , name="budget")
```

```python
# Objective function
obj = gp.quicksum(aij[commodity, category]*prices[commodity, category]
                                       for commodity in commodities for category in categories
                                          if  commodity in accra.loc[accra.category==category].commodity.unique().tolist())

consumption.setObjective(obj, GRB.MINIMIZE)
```

```python
consumption.optimize()
```

```python
def get_aub_line(i):
    
    aub1 = accra_prices.copy()
    
    aub1.loc[aub1.category==categories[i], "price"] = -1

    aub1.loc[aub1.category!=categories[i], "price"] = 0
    
    return aub1.pivot(index="category", columns="commodity", values="price").fillna(0).values.flatten()
```

```python
def get_aub_line2(i):
    
    aub1 = accra_prices.copy()
    
    aub1.loc[aub1.commodity==commodities[i], "price"] = -1

    aub1.loc[aub1.commodity!=commodities[i], "price"] = 0
    
    return aub1.pivot(index="category", columns="commodity", values="price").fillna(0).values.flatten()
```

```python
aub = np.array([get_aub_line(i) for i in range(4)]).reshape((4, 88))
```
